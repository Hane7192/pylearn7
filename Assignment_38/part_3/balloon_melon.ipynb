```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10
plt.rcParams['legend.fontsize'] = 10
plt.rcParams['axes.labelsize'] = 10
plt.rcParams['axes.titlesize'] = 15
```

```python
N = 500

balloons = pd.DataFrame({
    'length': np.random.normal(10, 4, N),
    'width': np.random.normal(25, 4, N),
    'weight': np.random.normal(10, 8, N)})

melons = pd.DataFrame({
    'length': np.random.normal(15, 3, N),
    'width': np.random.normal(15, 3, N),
    'weight': np.random.normal(1000, 200, N)})
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter3D(balloons['length'], balloons['width'], balloons['weight'], label='Balloons', alpha =0.5, color= 'OrangeRed')
ax.scatter3D(melons['length'], melons['width'], melons['weight'], label= 'Melons', alpha = 0.5, color= 'LightSeaGreen')

plt.title("Balloons & Melons")
ax.set_xlabel("Length (cm)")
ax.set_ylabel("Width (cm)")
ax.set_zlabel("Weight (gr)")
plt.legend(loc="upper left")

ax.view_init(30, 60)
plt.show()
```
